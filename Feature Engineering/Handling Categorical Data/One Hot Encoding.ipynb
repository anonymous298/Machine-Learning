# **OneHotEncoding Using Pandas**

```python
one_hot_encoded_df = pd.get_dummies(df, columns=['fuel', 'owner'])

one_hot_encoded_df = one_hot_encoded_df.map(lambda x: int(x) if isinstance(x, bool) else x)
```
