```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv('HR_comma_sep.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
X = dataset.satisfaction_level
y = dataset['left']
```

```python
plt.scatter(X,y, color='r')
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset[['satisfaction_level']],y, test_size=0.1)
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
print(f"Correlation: {model.coef_}")
```

```python
sns.barplot(x=model, y=X_test)
```
